```python
from transformers import AutoTokenizer

# 1. Hugging Face üzerinden BERT'in çok dilli tokenizer modelini indiriyoruz
tok = AutoTokenizer.from_pretrained('bert-base-multilingual-cased')

# 2. Üzerinde test yapacağımız örnek Türkçe cümlemiz
metin = 'Yapay zeka ile makine öğrenmesi çok eğlenceli.'

# 3. Çıktıları ekranda incelemek için print komutlarımızı basıyoruz
print("1. Token Parçaları:")
print(tok.tokenize(metin))          # Metnin alt parçalara bölünmüş hali

print("\n2. Token ID Karşılıkları:")
print(tok.encode(metin))            # Her bir parçanın sözlükteki ID numarası

print("\n3. Toplam Token Sayısı:")
print(len(tok.encode(metin)))       # Cümlenin toplamda kaç token harcadığı
```

```python
from sentence_transformers import SentenceTransformer

print("Model yükleme işlemi başlatılıyor, lütfen bekleyin...")

# Güvenli ve sıfırdan bir indirme bağlantısı tetikliyoruz
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# Örnek cümlemizi encode ediyoruz
vec = model.encode('Bugün hava çok güzel.')

print("\n=== EMBEDDING BAŞARIYLA ÜRETİLDİ ===")
print("1. Vektörün Boyutu (Kaç sayıdan oluşuyor?):")
print(vec.shape)      

print("\n2. Bu Cümle Vektörünün İlk 10 Sayısı (Anlam Örüntüsü):")
print(vec[:10])
```

```python
# Biri tek kelimelik, diğeri upuzun iki farklı metin tanımlıyoruz
kisa_metin = "Kedi."
uzun_metin = "Yapılan son araştırmalara göre, evcil kedilerin günlük uyku süreleri ve beslenme alışkanlıkları doğal ortamlarındaki akrabalarından oldukça farklılık göstermektedir."

# İki metni de embedding modeline sokuyoruz
vec_kisa = model.encode(kisa_metin)
vec_uzun = model.encode(uzun_metin)

print("=== SABİT BOYUT DOĞRULAMA TSETİ ===")
print("Kısa Metnin Vektör Boyutu (Shape):", vec_kisa.shape)
print("Uzun Metnin Vektör Boyutu (Shape):", vec_uzun.shape)
```

```python
from sentence_transformers import util

# Üç farklı cümleyi embedding modelimizle uzaya gönderiyoruz
a = model.encode('Kedi bahçede uyuyor.')
b = model.encode('Bir pisi çimenlerde şekerleme yapıyor.')
c = model.encode('Borsa bugün yüzde üç düştü.')

print("=== ANLAMSAL BENZERLİK SONUÇLARI ===")

# Kedi ve Pisi cümleleri arasındaki Kosinüs Benzerliğini ölçüyoruz
sim_ab = util.cos_sim(a, b).item()
print("Kedi cümlesi ile Pisi cümlesi benzerlik skoru:")
print(f"{sim_ab:.4f} (Anlamca yakın çıkmalı)")

# Kedi ve Borsa cümleleri arasındaki Kosinüs Benzerliğini ölçüyoruz
sim_ac = util.cos_sim(a, c).item()
print("\nKedi cümlesi ile Borsa cümlesi benzerlik skoru:")
print(f"{sim_ac:.4f} (Alakasız çıkmalı)")
```

```python
# (a) Anlamca aynı ama kelimeler tamamen farklı çift
a1 = model.encode("Hekim hastayı muayene etti.")
a2 = model.encode("Doktor tıp merkezinde hastaya baktı.")

# (b) Aynı kelimeleri içeren ama anlamı tamamen farklı çift (Zor Test)
b1 = model.encode("Köpek kediyi kovaladı.")
b2 = model.encode("Kedi köpeği kovaladı.")

print("=== ÖZEL DENEY SONUÇLARI ===")
print("(a) Farklı Kelimeler / Aynı Anlam Skoru:", util.cos_sim(a1, a2).item())
print("(b) Aynı Kelimeler / Farklı Anlam Skoru:", util.cos_sim(b1, b2).item())
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Silinen orijinal cümle listesini ve bir önceki adımdaki modeli kullanarak V'yi yeniden üretiyoruz
cumleler = [
    'Kedi bahçede uyuyor.',
    'Bir pisi çimenlerde şekerleme yapıyor.',
    'Köpek parkta koşuyor.',
    'Borsa bugün yüzde üç düştü.',
    'Dolar kuru sabah yükseldi.'
]
V = model.encode(cumleler)

# 1. K-Means ile Gözetimsiz Kümeleme yapıyoruz (Hayvanlar vs. Ekonomi için n_clusters=2)
kume = KMeans(n_clusters=2, random_state=42).fit_predict(V)
print("=== CÜMLELERİN KÜME NUMARALARI ===")
print(kume)      # Her cümlenin atandığı küme ID'sini (0 veya 1) ekrana basar

# 2. PCA ile 384 boyutlu embedding vektörlerini grafikte çizebilmek için 2 boyuta indiriyoruz
V2 = PCA(n_components=2).fit_transform(V)

# 3. İndirgenen 2 boyutlu koordinatları matplot ile ekrana saçıyoruz (Renkler küme ID'sine göre)
plt.scatter(V2[:, 0], V2[:, 1], c=kume, cmap='coolwarm', s=200)

# 4. Döngü ile her noktanın hemen yanına ilgili cümlenin ilk 20 karakterini etiket olarak yazıyoruz
for i, c in enumerate(cumleler):
    plt.annotate(c[:20], (V2[i, 0], V2[i, 1]), xytext=(5, 5), textcoords='offset points')

# 5. Grafiği ekranda gösteriyoruz
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# 1. Model ve Cümleleri tekrar tanımlıyoruz
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
cumleler = [
    'Kedi bahçede uyuyor.',
    'Bir pisi çimenlerde şekerleme yapıyor.',
    'Köpek parkta koşuyor.',
    'Borsa bugün yüzde üç düştü.',
    'Dolar kuru sabah yükseldi.'
]
V = model.encode(cumleler)

# 2. PCA ile 2 boyutlu koordinatları tek seferde hesaplıyoruz
V2 = PCA(n_components=2).fit_transform(V)

# 3. n_clusters değerlerini (2, 3, 4) döngü ile test ediyoruz
for k in [2, 3, 4]:
    kume = KMeans(n_clusters=k, random_state=42).fit_predict(V)
    
    plt.figure(figsize=(6, 4))
    plt.scatter(V2[:, 0], V2[:, 1], c=kume, cmap='rainbow', s=200)
    
    for i, c in enumerate(cumleler):
        plt.annotate(c[:20], (V2[i, 0], V2[i, 1]), xytext=(5, 5), textcoords='offset points')
        
    plt.title(f"K-Means Kümelemesi (n_clusters = {k})")
    plt.grid(True)
    plt.show()
    print(f"k={k} için Küme Dağılımları: {kume}\n" + "-"*50)
```
